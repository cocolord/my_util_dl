```
import cv2

target_path = "/content/drive/My Drive/dataset_918/iPhone"

def video2frame(videos_path,frames_save_path,time_interval=1):
 
  '''
  :param videos_path: 视频的存放路径
  :param frames_save_path: 视频切分成帧之后图片的保存路径
  :param time_interval: 保存间隔
  :return:
  '''
  vidcap = cv2.VideoCapture(videos_path)
  success, image = vidcap.read()
  count = 0
  while success:
    try:
      success, image = vidcap.read()
      count += 1
      if count % time_interval == 0:
        cv2.imencode('.png', image)[1].tofile(frames_save_path + "/%d.png" % count)
    except:
      pass
  print(count)

file_list = os.listdir(target_path)
print(file_list)
for file in file_list:
  if os.path.splitext(file)[-1] == '.MOV':
    frames_save_path = os.path.join(target_path,file.split('.')[0])
    print(frames_save_path)
    try:
      os.mkdir(frames_save_path)
    except:
      pass
    try:
      os.mkdir(os.path.join(frames_save_path,'output'))
    except:
      pass
    video2frame(os.path.join(target_path,file), frames_save_path)
```
